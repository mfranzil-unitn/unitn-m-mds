# WORKING WITH IMAGES

There are several different libraries available for processing and manipulating images in Python. Among the most common we can list:

*   [skimage](https://scikit-image.org/)
*   [Pillow](https://pillow.readthedocs.io/en/stable/)
*   [OpenCV](https://www.pyimagesearch.com/2018/07/19/opencv-tutorial-a-guide-to-learn-opencv/)


```python
import cv2
import matplotlib.pyplot as plt
```

**Open** an image and display it

```python
im = cv2.imread('../Images/test.jpg')
print('Value of top-left pixel: ' + str(im[0, 0, :]))

cv2.imshow('Test', im)
# cv2_imshow(im)
```

Extract a color channel and display it

```python
# Plot the 3 separate components
# Pay attention, in openCV images use BGR istead of RGB
plt.subplot(131)
plt.title('R')
plt.imshow(im[:, :, 2], cmap='gray')
plt.subplot(132)
plt.title('G')
plt.imshow(im[:, :, 1], cmap='gray')
plt.subplot(133)
plt.title('B')
plt.imshow(im[:, :, 0], cmap='gray')
plt.show()
```

Convert color space

```python
# Pay attention, openCV uses YCrCb instead of YCbCr

ycrcb = cv2.cvtColor(im, cv2.COLOR_BGR2YCR_CB)
print(type(ycrcb))

# We cast the transformed values in 3x8 bits per pixel to allow visualization
# cv2_imshow(ycrcb)
cv2.imshow("Transformed values", ycrcb)

print('Value of top-left pixel: ' + str(ycrcb[0, 0, :]))

plt.figure(figsize=(15, 3))

# Plot the 3 separate components
plt.subplot(131)
plt.title('Y')
plt.imshow(ycrcb[:, :, 0], cmap='gray')
plt.subplot(132)
plt.title('Cb')
plt.imshow(ycrcb[:, :, 2], cmap='gray')
plt.subplot(133)
plt.title('Cr')
plt.imshow(ycrcb[:, :, 1], cmap='gray')
plt.show()
```

Save a transformed image as JPEG

```python
cv2.imwrite('transformed.jpg', ycrcb)
```
